```python
import requests
from bs4 import BeautifulSoup
from PIL import Image
from io import BytesIO
import os

def get_image_urls(query, num_images = 5):
    search_url = f"https://www.google.com/search?1={query}&source=lnms&tbm=isch"
    headers = {"User-Agent": "Mozilla/5.0"}
    response = requests.get(search_url, headers = headers)
    soup = BeautifulSoup(response.text, 'html.parser')

    image_urls = []
    for img in soup.find_all('img', limit=num_images+1):
        url = img.get('src')
        if url and 'http' in url:
            image_urls.append(url)
    return image_urls

def download_images(image_urls, download_path):
    if not os.path.exists(download_path):
        os.makedirs(download_path)

    for i, url in enumerate(image_urls):
        try:
            response = requests.get(url)
            img = Image.open(BytesIO(response.content))

            img_filename = os.path.join(download_path, f"image_{i+1}.jpg")
            img.save(img_filename)
        
        except Exception as e:
            print(f"Could not download{url}: {e}")
```
